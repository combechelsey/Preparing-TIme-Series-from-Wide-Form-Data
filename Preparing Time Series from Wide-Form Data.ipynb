```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
# set random seed
SEED = 321
np.random.seed(SEED)

## Customization Options
# plt.style.use(['fivethirtyeight'])#'seaborn-talk'])#,
plt.rcParams['figure.facecolor']='white'
plt.rcParams['figure.figsize']=(12,4)
#import tick customization tools
import matplotlib.ticker as mticks
import matplotlib.dates as mdates
sns.set_context("talk", font_scale=0.9)

def format_xdates(ax, fig=None, xtick_fontweight='bold',
               title= None):
    """Formats x-axis with minor ticks every 3 months, and major 
    ticks every year. Includes different grid options for visibility"""
    # create the locator to place ticks every 3 months.
    loc_3months = mdates.MonthLocator(interval=3)
    fmt_months = mdates.DateFormatter("%b")
    ## for major year ticks
    loc_year = mdates.YearLocator()
    fmt_year = mdates.DateFormatter("%Y")
    
    ## Change the font of the major ticks to stand out
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), 
                       fontweight=xtick_fontweight) 
    
    ## customize minor ticks
    ax.xaxis.set_minor_locator(loc_3months)
    ax.xaxis.set_minor_formatter(fmt_months)
    ## customize major ticks
    ax.xaxis.set_major_locator(loc_year)
    ax.xaxis.set_major_formatter(fmt_year)
    ## Making major/minor gridlines visually distince
    ax.grid(which='minor',axis='x',ls=":")
    ax.grid(which='major',axis='x',color='k')
    if fig is None:
        fig = ax.get_figure()
    ## rotate the dates
    fig.autofmt_xdate(which='major',rotation=90,ha='center')
    return fig,ax
```

```python
# Load Data
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv"
df = pd.read_csv(url)
df
```

- Filter the 4 largest cities into a new dataframe.

    - Tip: the "SizeRank" column has already ranked the cities by size. The larger the city, the smaller the rank value.
        - Therefore the 4 largest cities would have rank values of [0,1,2,3]

```python
# Filter for size rank < 4
df = df[df['SizeRank']<4]
```

```python
# Checking that filter was applied correctly
df["SizeRank"].value_counts()
```

- Melt the data to long-form and prepare it for time series analysis.

    - Convert the melted dates into datetime datatype.
    - Make the datetime column the index.

```python
# Combine date columns
melted = pd.melt(df, id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName'], var_name="Date")
melted
```

```python
# Changing date column to datetime type
melted['Date'] = pd.to_datetime(melted['Date'])
```

```python
# Setting Date as column index
melted = melted.set_index("Date")
```

```python
# Viewing Dataframe 
melted
```

- Resample the dataframe as monthly frequency, grouped by City.

```python
# Resample as monthly frequency grouped by city
city_month = melted.groupby('RegionName').resample("M").sum()
city_month


```

```python
value_trends = unstacked['value']
```

```python
from matplotlib.ticker import FuncFormatter

fig, ax = plt.subplots(figsize=(15,8))

# Set labels and title
ax.set_xlabel("Date")
ax.set_ylabel("Home Values ($)")
ax.set_title("Average Home Values by City")

# Create a function to format the y-axis ticks
def format_thousands(value, tick_number):
    return f'{int(value/1000)}K'

# Apply the custom formatter to the y-axis
ax.yaxis.set_major_formatter(FuncFormatter(format_thousands))

# create the locator to place ticks every 3 months.
loc_3months = mdates.MonthLocator(interval=3)
fmt_months = mdates.DateFormatter("%b")
## for major year ticks
YearLocator = mdates.YearLocator()
fmt_year = mdates.DateFormatter("%Y")

# Customize the major x-axis ticks
loc_year = mdates.YearLocator(base=1)  # Major ticks every 1 year
fmt_year = mdates.DateFormatter('%Y')  # Format as 4-digit years

## customize major ticks
ax.xaxis.set_major_locator(loc_year)
ax.xaxis.set_major_formatter(fmt_year)

## customize minor ticks
ax.xaxis.set_minor_locator(loc_3months)
ax.xaxis.set_minor_formatter(fmt_months)

## Making major/minor gridlines visually distinct
ax.grid(which='minor',axis='x',ls=":")
ax.grid(which='major',axis='x',color='k')
fig.autofmt_xdate(rotation=90 ,ha='center')

ax.plot(value_trends.index, value_trends)


# Remove the labels for minor ticks
ax.tick_params(axis='x', which='minor', labelbottom=False)
ax.legend(value_trends)


# Show the plot
plt.tight_layout()
plt.show(); 
```

- 2) How much did the home values change from November 2008 to December 2008 (in dollars)?

    - Hint: you can use .diff() to calculate the change in values

```python
# Viewing Dataframe 
melted
```

```python
# Resample as monthly frequency grouped by city
city_month_ave = melted.groupby('RegionName')['value'].resample("M").mean()
city_month_ave
```

```python
# Unstacked city_month_ave
unstacked = city_month_ave.unstack(level=0)

unstacked
```

```python
# .diff()
change = city_month_ave.diff()
change
```

```python
# Find change for dec 2008

unstacked_diff = unstacked.diff()
change = unstacked_diff.loc["12-2008"]

change
```
